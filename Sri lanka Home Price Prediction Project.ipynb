```python
import pandas as pd
```

```python
df = pd.read_csv('house_prices.csv')
df.head(3)
```

```python
df1 = df.drop(df[['Title','Sub_title','Address','Description','Post_URL','Seller_name','Seller_type','published_date','Geo_Address','Lat','Lon']],axis='columns')
df1.head()
```

```python
df1.isnull().sum()
```

```python
#df1.Location.unique()
df1['town'] = df1.Location.apply(lambda x: str(x).split(',')[0])
df1.head()
```

```python
df1 = df1.drop('Location', axis='columns')
df1.head()
```

```python
df1['Price'] = df1.Price.str.replace('Rs', '').str.replace(',', '')
df1.head()
```

```python
df1['Land_size'] = df1.Land_size.apply(lambda x: str(x).replace(',', '').split()[0])
df1.head()
```

```python
df1['House_size'] = df1.House_size.apply(lambda x: str(x).replace(',', '')[0])
df1.head()
```

```python
df1.Beds.unique()
```

```python
df1.town
df1['town'] = df1['town'].str.strip()
```

```python
df1.shape
```

```python
dummies = pd.get_dummies(df1.town)
dummies
```

```python
df2 = pd.concat([df1,dummies],axis='columns')
df2.columns
```

```python
df3 = df2.drop('town', axis='columns')
df3.head()
df3 = df3[df3['Baths'] != '10+']
df3['Baths'] = df3.Baths.astype(float)
df3.Baths.unique()
```

```python
df3 = df3[df3['Beds'] != '10+']
df3['Beds'] = df3.Beds.astype(float)
df3.Beds.unique()
```

```python
X = df3.drop('Price', axis='columns')
X[:10]
```

```python
y = df3.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
bg = BaggingRegressor()
rf = RandomForestRegressor()
svr = SVR()
get_score(lr, X_train, X_test, y_train, y_test)
```

```python
def predict_price(Baths, Land_size, Beds, House_size, location):
    try:
        loc_index = np.where(X.columns == location)[0][0]
    except IndexError:
        # Handle case where location is not found
        print(f"Location '{location}' not found in X.columns")
        return None
    
    res = np.zeros(len(X.columns))
    res[0] = Baths
    res[1] = Land_size
    res[2] = Beds
    res[3] = House_size
    
    if loc_index >= 0:
        res[loc_index] = 1
        
    return rf_model.predict([res])[0]
```
